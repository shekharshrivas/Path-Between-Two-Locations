```python
def BFS_Path(graph, start_node, target_node):
    """ return the path list from start_node to target_node or return empty list
        if path does not exist.
         
        test case:
        >>>BFS_path(graph, Agra, delhi)
        ['Agra', 'Bikaner', 'Delhi']
        
        parameter graph: adjacency list of all the random connected graph
        type graph: dict
         
        parameter start_node: starting city
        type start_node: str
         
        parameter target_node: goal city
        type target_node: str
        """

    visited = []
    queue = [(start_node, [start_node])]

    while queue:
        node, path = queue.pop(0)

        if node == target_node:
            return path  # Return the path if the target node is found

        if node not in visited:
            visited.append(node)

        for neighbor in graph.get(node, []):
            if neighbor not in [x for x, y in queue] and neighbor not in visited:
                new_path = path + [neighbor]
                queue.append((neighbor, new_path))

    return []  # If the target node is not found, return an empty list

start_node = "Agra"
target_node = "Delhi"

shortest_path = BFS_Path(adjacency_list, start_node, target_node)

if shortest_path:
    print(f"Shortest path from {start_node} to {target_node}: {shortest_path}")
else:
    print(f"No path found from {start_node} to {target_node}.")
```

```python
# This code is written by with the help of ChatGPT.
# But I made some changes in code. So that it will work fine.

def Astar(graph, start_node, goal_node, route_value, heuristic_value):
    
    """
    return the path list from start_node to goal_node if exist otherwise None.

    test case:
    >>>Astar(graph, Agra, Kota, route_value, heuristic_value)
    ['Agra', 'Bundi', 'Patna', 'Nawada', 'Kota']

    parameter graph: adjacency list of all the random connected graph.
    type graph: dict

    parameter start_node: starting city.
    type start_node: str
    
    parameter goal_node: starting city.
    type goal_node: str

    parameter route_value: route_value is the function which gives actual route distance from node1 to node2.

    parameter heuristic_value: heuristic_value is the function which gives heuristic distance from node1 to node2.
    """
    open_list = [start_node]
    came_from = {}  # Initialize a dictionary to track the path
    g_score = {node: float('inf') for node in graph}
    g_score[start_node] = 0  # The cost from start to start is 0
    visited_list = []

    while open_list:
        # Find the node in the open list with the lowest combined cost
        current = min(open_list, key=lambda node: g_score[node] + heuristic_value(node, goal_node))

        if current == goal_node:
            path = []
            while current in came_from:
                path.insert(0, current)  # Reconstruct the path from goal to start
                current = came_from[current]
            path.insert(0, start_node)
            return path  # Return the path when the goal is reached

        open_list.remove(current)
        visited_list.append(current)

        for neighbor in graph[current]:
            tentative_g_score = g_score[current] + route_value(current, neighbor)

            if tentative_g_score < g_score[neighbor]:
                came_from[neighbor] = current  # Record the path to this neighbor
                g_score[neighbor] = tentative_g_score  # Update the cost
                if neighbor not in open_list and neighbor not in visited_list:
                    open_list.append(neighbor)  # Add the neighbor to the open list
    return None  # No path found


# Example usage:
start_node = "Agra"
goal_node = "Kota"

# Perform A* search
path = Astar(adjacency_list, start_node, goal_node, route_value, heuristic_value)


if path:
    print("A* Path found::", " -> ".join(path))
else:
    print("No path found.")
```
